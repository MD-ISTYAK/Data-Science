```
df=data

for col in data.columns:
    q25=df[col].quantile(0.25)
    q75=df[col].quantile(0.75)
    iqr=q75-q25
    max_val=q75+(1.5*iqr)
    min_val=q25-(1.5*iqr)

    # Use bitwise AND operator (&) to combine conditions, and parentheses for clarity
    df.loc[ (df[col]<min_val) & (df[col]>max_val), col]= " "

    print(col, df[col].isnull().sum())
```

```
x_data=data.drop('Personal Loan', axis=1)
y_data=data['Personal Loan']
```

```
x_data.shape
```

```
y_data.shape
```

```
# Create the training and test data set in the ratio of 70:30 respectively. Can be any other ratio...
# Prepare data for models
from sklearn import model_selection
features=x_data
target=y_data

test_size=0.20

x_train, x_test, y_train, y_test=model_selection.train_test_split(features, target, test_size=test_size, random_state=7)
```

```
x_train.shape
```

```
x_test.shape
```

```
####### --------------------------- Logistic Regression --------------------------------------
```

```
from sklearn.linear_model import LogisticRegression
from sklearn import metrics

model=LogisticRegression()
model.fit(x_train, y_train)
```

```
y_predict=model.predict(x_test)
model_score=model.score(x_test, y_test)
print(model_score)
```

```
print(metrics.confusion_matrix(y_test, y_predict))
```

```
print(metrics.classification_report(y_test, y_predict))
```

```
#The precision and recall for class 1 is low...

#Precision: Within a given set of positively-labeled results, the fraction that were
#true positives = tp/(tp + fp) , this has to be calculated for each class i.e. 0 and 1 and should be
# high for the class less represented, class 1 in our example

#Recall: Given a set of positively-labeled results, the fraction of all positives that were
#retrieved = tp/(tp + fn)

# Accuracy: tp + tn / (tp + tn + fp +fn) But this measure can be dominated by larger class.
# Suppose 10, 90 and 80 of 90 is correctly predicted while only 2 of 0 is predicted correctly.
# Accuracy is 80+2 / 100 i.e. 82%

# F is harmonic mean of precision and recal given by ((B^2 +1) PR) / (B^2P +R)
#When B is set to 1 we get F1 = 2PR / (P+R)
```

```
#####  ------------------------------ K Nearest Neighbours ------------------------------------
```

```
from sklearn.neighbors import KNeighborsClassifier

NNH=KNeighborsClassifier(n_neighbors=3)
NNH.fit(x_train, y_train)
```

```
predict_labels=NNH.predict(x_test)
NNH_score=NNH.score(x_test, y_test)
print(NNH_score)
```

```
print(metrics.confusion_matrix(y_test, predict_labels))
```

```
print(metrics.classification_report(y_test, predict_labels))
```

```
### Recall (true positives / (true positives + false negatives)) for class 1 is the least. That is
### because majority of data points belong to class 0 and in KNN, probability of finding data points
### from class 0 closer to a test point than a data point from class 1 is high.
```

```
##--let scale the x_train and x_test and see the changes in our models
```

```
from sklearn import preprocessing

x_train_scaled=preprocessing.scale(x_train)
x_test_scaled=preprocessing.scale(x_test)
```

```
#  Using K Nearest Neighbours with scaled values
```

```
NNH.fit(x_train_scaled, y_train)
predicted_label=NNH.predict(x_test_scaled)
NNH_score=NNH.score(x_test_scaled, y_test)
print(NNH_score)
```

```
print(metrics.confusion_matrix(y_test, predicted_label))
```

```
print(metrics.classification_report(y_test, predicted_label))
```

```
## scaled KNN algorithm gives the best result for class 1 till now with 89% precision
## model score is 95.6%
```

```
## Using K logistic regression with scaled values
```

```
model.fit(x_train_scaled, y_train)
y_scaled_predict=model.predict(x_test_scaled)
model_score=model.score(x_test_scaled, y_test)
print(model_score)
```

```
print(metrics.confusion_matrix(y_test, y_scaled_predict))
```

```
print(metrics.classification_report(y_test, y_scaled_predict))
```

```
## scaled KNN algorithm gives the good result for class 1 till now with 83% precision
## model score is 95.1%
```
